## Hopkins Test 
#### Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express no clusterability, a score tending to 1 express an uniform distrubution so clustering can't be useful and  0 express  not an uniform distrubution hence clusterering can be useful.
***********
#### Bir veri kümesinin kümelenebilirliğini değerlendirin. 0 ile 1 arasında bir puan, 0,5 civarında bir puan kümelenebilirlik olmadığını ifade eder, 1'e meyilli bir puan tek tip bir dağılımı ifade eder, bu nedenle kümeleme yararlı olamaz ve 0 tekdüze bir dağılımı ifade etmez, dolayısıyla kümeleme yararlı olabilir.

```python
from pyclustertend import hopkins
```

```python
df_ = df.copy()
```

```python
hopkins(df_, df.shape[0])  

# 0 ile 0.5 arasinda degerler gelmesini bekleriz. Sifira ne kadar yakin olursa kümelemeye o kadar yatkin demektir.
```

```python
K_means_model.fit_predict(X)
```

### Elbow Method 
### We use this method to find optimal k value by looking up to break point of graph. 
*********
### Grafiğin kırılma noktasına bakarak optimal k değerini bulmak için bu yöntemi kullanırız.

```python
ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method") 
```

* Keskin düsüsün durdugu ilk nokta 3 tür. Her zaman ilk noktayi aliyoruz.

```python
ssd
```

```python
pd.Series(ssd).diff()
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

```python
df_diff.plot(kind='bar') # keskin düsüsün durdugu ilk nokta, 3 noktasi oldugu asikar. Buna göre k=3.
```

```python
from yellowbrick.cluster import KElbowVisualizer

visualizer = KElbowVisualizer(KMeans(random_state=42), k=(2,9))
visualizer.fit(X_scaled)        # Fit the data to the visualizer
visualizer.show();
```

*  k=4 oldu. Elbow Metod ve Yellow Brick yaklasimlari 100% degildir. Bu nedenle Silhouette analizine de bakiyorum.

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_scaled, KMeans(random_state = 42).fit_predict(X_scaled))
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    silhouette_ = silhouette_score(X_scaled, 
                                      KMeans(n_clusters=num_clusters, random_state=42).fit_predict(X_scaled))
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_}")
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer

visualizer = SilhouetteVisualizer(KMeans(n_clusters=4, random_state=42))
visualizer.fit(X_scaled)    # Fit the data to the visualizer
visualizer.poof();

# Burada Elbow method ve görsellere de bakip 3 e karar veriyoruz. Daha iyi kümelenmenin insight ini veriyor.
```
